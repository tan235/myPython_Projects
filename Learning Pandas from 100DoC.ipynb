# Calculating the difference between to column and assigning it to new column

```python
spread_c = clean_ds['Mid-Career 90th Percentile Salary'] - clean_ds['Mid-Career 10th Percentile Salary']
clean_ds.insert(1, 'SSpread', spread_c)
clean_ds.head()
```

Finding the low risk data

```python
low_risk = clean_ds.sort_values('SSpread')
low_risk[['Undergraduate Major','SSpread']].head() 
```

```python
low_risk[['Undergraduate Major','SSpread']].tail() 
```

Using the .sort_values() method, can you find the degrees with the highest potential? Find the top 5 degrees with the highest values in the 90th percentile. 

```python
top_5 = clean_ds.sort_values('Mid-Career 90th Percentile Salary', ascending = False)
top_5[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

Also, find the degrees with the greatest spread in salaries. Which majors have the largest difference between high and low earners after graduation.

```python
high_pot = clean_ds.sort_values('SSpread', ascending = False)
high_pot[['Undergraduate Major','SSpread']].head() 
```

Learning groupby() method

```python
clean_ds.groupby('Group').count()
```

Now can you use the .mean() method to find the average salary by group? 

```python
pd.options.display.float_format = '{:,.2f}'.format  # adds , for better readability
clean_ds.groupby('Group').mean()
```
